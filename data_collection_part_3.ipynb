```python
import os
import pandas as pd
```

```python
tweet_count = 500000
text_query = "asgari ücret"
since_date = "2022-10-15"
until_date = "2023-02-13"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> minwagepr5.json'.format(tweet_count, since_date, text_query, until_date))

# Reads the json generated from the CLI command above and creates a pandas dataframe
min_wage = pd.read_json('minwagepr5.json', lines=True)

# Displays first 5 entries from dataframe
min_wage
```
