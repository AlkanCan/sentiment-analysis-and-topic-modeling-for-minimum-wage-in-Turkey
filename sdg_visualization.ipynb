```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sdg_1=pd.read_csv('minwagesentiment1.csv')
```

```python
sdg_1
```

```python
sdg_2=pd.read_csv('minwagesentiment2.csv')
```

```python
sdg_2
```

```python
sdg_3=pd.read_csv('minwagesentiment3.csv')
```

```python
sdg_3
```

```python
df=pd.concat([sdg_2,sdg_1,sdg_3],ignore_index=True)
```

```python
df
```

```python
df=df.drop_duplicates(subset=['id'])
```

```python
df=df[['date','rawContent','labels']]
```

```python
df
```

```python
df=df.sort_values(by=['date'])
```

```python
df=df.reset_index(drop=True)
```

```python
df
```

```python
df.to_csv('sdgwholedatasetWsentiment.csv')
```

```python
df=pd.read_csv('sdgwholedatasetWsentiment.csv')
```

```python
plt.pie(df['labels'].value_counts(),labels=['Negative','Positive'],autopct='%1.1f%%')
plt.show()
df['labels'].value_counts()
```

```python

#convert the 'Date' column to datetime format
df['date']= pd.to_datetime(df['date'])
 
# Check the format of 'Date' column
df.info()
```

```python
df.resample('d', on='date').size().plot(xlabel ='Month of year', 
                                         ylabel='Tweets',
                                         figsize=(17,6),
                                         grid=True,
                                        colormap="tab10",
                                        title="Number of tweets sent every day")
```

```python
df_pos=df[df['labels']=='positive']
```
