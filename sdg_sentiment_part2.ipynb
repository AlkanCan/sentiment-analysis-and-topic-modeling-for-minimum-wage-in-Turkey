```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('minimumwagepart1.csv')
```

```python
df
```

```python
tweets=df['rawContent']
```

```python
tweets
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline
```

```python
model = AutoModelForSequenceClassification.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
sa= pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)
```

```python
sentiment=[]
```

```python
for i in range(len(tweets)):
    if i % 10_000 == 0:
        print(f"{i}/{len(tweets)} | {i/len(tweets)*100:.2f}%")
    sentiment.append(sa(tweets[i]))

print(f"{len(tweets)}/{len(tweets)} | {len(tweets)/len(tweets)*100:.2f}%")
```

```python
sentiment
```

```python
labels=[i[0]['label'] for i in sentiment]
```

```python
labels
```

```python
labels_series=pd.Series(labels,name='labels')
```

```python
labels_series
```

```python
df_wsentiment=pd.concat([df,labels_series],axis=1)
```

```python
df_wsentiment
```

```python
df_wsentiment.to_csv('minwagesentiment2.csv')
```
